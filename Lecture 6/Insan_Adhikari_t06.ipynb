Create a function that will perform the following calculation on a given input value and return the value of y: 

$y = 5 \sin\left(\frac{x-6\pi}{12}\right)$

(The input value here is $x$, and $\pi = 3.14159...$ is a constant.)

```python
import numpy as np

def f(x):
    return 5 * np.sin((x-6*np.pi)/12)
print(f(6 * np.pi)) # should return 0    
print(f(0)) # should return -5  
```

Complete the loop below to create a list named "x_list" that contains values ranging from 0 to 100, incremented by 0.25 (i.e.; 0, 0.25, 0.5, ... 99.5, 99.75, 100). Print out the list after creating it.

```python
x_list = []
count = 0
while count <= 100:
    x_list.append(count)
    count += 0.25
    
print(x_list)
```

Now create "y_list", which contains the value $y$ as defined in your function, given the $x$ values from "x_list". Then, print the first value of "y_list".

```python
y_list = []
for x in x_list:
    y_list.append(f(x))

print(y_list)
```

Run the cell below to plot "x_list" against "y_list".

```python
import matplotlib.pyplot as plt

plt.plot(x_list,y_list)
plt.xlabel('x')
plt.ylabel('y')
```

Create a new function called "y_from_list" that takes in an entire list of $x$ values as input, and then calculates all of the corresponding $y$ values and returns a list of those $y$ values.

```python
def y_from_list(input_list):
    output_list = []
    for input in input_list:
        output_list.append(f(input))
    return output_list
```
