# Luminosity

Luminosity measures the total energy per second generated by a star [Joules/sec]. It depends on the surface area of the star, and its temperature. 

L = 4 $\pi R^2$ $\cdot$ $\sigma T^4$

$\sigma$ is the Stefan–Boltzmann constant, with a value of $5.67\times10^{−8} W⋅m^{−2}\cdot K^{−4}$.

Let's write a function that calculates the luminosity of a star for a given Temperature and a given Radius. 

```python
sigma = 5.67e-8    # Stefan-Boltzman constant
T = 5778.0        # Temperature of the Sun in Kelvin
Rsun = 6.96e8    # Radius of the Sun in meters

from math import pi

def luminosity1(T):
    L = 4 * pi * Rsun**2 * sigma* T**4

    return L

print(luminosity1(T))

print("Luminosity of a Sun-size star with Temperature {:.0f} K is {:.3g} W".format\
      (T, luminosity1(T)))
```

Next we'll create a list of temperatures at which to evaluate Luminosity:

```python
Tlist = [] 
# Make a list that ranges from 3,000 to 12,000 K
for i in range(10):
    T = 3000 + 1000*i
    Tlist.append( T )
print(Tlist)
```

Now, we'll create a Luminosity list and call the "luminosity1" function to evaluate it:

```python
luminosityList = []
# loop over list of temperatures
for T in Tlist:
    L = luminosity1(T) # evaluate the luminosity at temperature T
    luminosityList.append(L)
print(luminosityList)
```

```python
for i in range(len(Tlist)):
    print("Temperature = {:}   Luminosity = {:.3g} ".format(Tlist[i],luminosityList[i]))
```

### Functions with multiple return values

```python
sigma = 5.67e-8    # Stefan-Boltzman constant
Tsun = 5778        # Temperature of the Sun in Kelvin
Rsun = 6.96e8    # Radius of the Sun in meters

from math import pi

def luminosity3(R,T):
    T_kelvin = T*Tsun
    R_meters = R*Rsun
    L = 4 * pi * (R_meters)**2 * sigma* T_kelvin**4

    return L, T_kelvin

# notice that the function is returning 2 things: Luminosity and Temperature

print(luminosity3(1, 1))
```

```python
result = luminosity3(1,1)
print(result)
print(type(result))
```

```python
lum, temper = luminosity3(1,1)
print(lum, temper)
```

# Doc Strings:

Doc strings are used to document functions.  They generally include:

* A description of the functionality of the function

* A list of arguments

* A description of outputs (returned values)

And, they serve as the help documentation!

They go right after the function header and are enclosed within triple quotes.

```python
def force(mass1, mass2, radius):
    """
    Compute the gravitational force between two bodies.
    
    Input Parameters
    ----------
    mass1 : int, float
        Mass of the first body, in kilograms.
    mass2 : int, float
        Mass of the second body, in kilograms.
    radius : int, float
        Separation of the bodies, in meters.

    Example
    -------
    To compute force between Earth and the Sun:
    >>> F = force(5.97e24, 1.99e30, 1.5e11)

    Returns
    -------
    Force in Newtons : float
    """
    G = 6.67e-11
    
    return G * mass1 * mass2 / radius**2
```

```python
result = force(5.97e24, 2.00e30, 1.5e11)
print(result)
```

```python
# To see the documentation for a function, use help:

help(force)
```

```python
from math import pi, exp, sin

# t is positional argument, others are keyword arguments
def f(t, A=1, a=1, omega=2*pi):   
    return A*exp(-a*t)*sin(omega*t)

v1 = f(0.01)  # Only the time is specified
print(v1)
```

### Nested functions: functions as arguments to other functions

$\frac{df(x)}{dx} \approx \frac{f(x-h) + f(x+h) - 2f(x)}{h^2}$

```python
def diff2(f, x, h=1e-6):
    """
    Calculates a second derivative.
    f:  the function (of one variable) to be differentiated
    x:  value at which the function is evaluated
    h:  small difference in x 
    """
    r = (f(x-h) + f(x + h) - 2.0 * f(x)) / float(h**2)
    return r
```

We'll work with the following function:

$g(t) = t^2$

```python
def g(t):
    return t**2
```

```python
t = 3.0
gPrimePrime = diff2(g, t)
```

```python
print("Second derivative of g=t^2 evaluated at t={:g}".format(t))
print("g({:f})={:.8g}" .format(t, gPrimePrime))
```
