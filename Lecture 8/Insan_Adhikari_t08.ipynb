# Tutorial #8: Numpy and Arrays

Create an array called "x_array" that ranges from 0 to 10 (0 and 10 included) with a step size of 0.25 and print it.

```python
import numpy as np
x_array = np.arange(0,10.25,0.25)
print(x_array)
```

Create a function named "y" that performs the following calculation on the input variable (which is $x$ in this example): $y = x^3 - 2x$.

```python
def y(x):
    return x**3 - 2*x
```

Complete the code below to make a plot of this function as a line on the interval [0,10] by using the function you've defined. 

Then plot every other element of the x_array and y_array datapoints as circles. 

Label your axes $x$ and $y$

```python
import matplotlib.pyplot as plt
y_array = y(x_array)

plt.plot(x_array, y_array, label='y = x^3 - 2x')
plt.scatter(x_array[::2], y_array[::2], color='red', label='Data points')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```

Create an array of random floats with a length of 1,000

Then find the following of the array: 

- It's largest and smallest values
- The average of all the elements in the array
- The sum of all the elements
- The sum of all the elements with indexes 100 to 120 (including both 100 and 120)

```python
rand_floats = np.random.rand(1000)
print(np.max(rand_floats))
print(np.min(rand_floats))
print(np.sum(rand_floats))
print(np.mean(rand_floats[100:121]))
```
