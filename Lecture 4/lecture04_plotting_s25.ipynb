# Plots

Python is a great programming langauge for data visualization! We can create line graphs from lists of data using the matplotlib.pyplot library (which we will import as plt).

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = [1, 2, 3,  4,  5,   6] # x-axis
y = [4, 8, 12, 16, 20, 24] # y-axis
```

Think of these two lists as sets of x and y coordinates. To plot these, simply use the plot method as shown below. The first argument in the plot method is the list of x-axis values and the second argument is the list of y-axis values. 

```python
plt.plot(x, y)
```

### Numpy Sidenote

The numpy library can be used to make certain list operations easier! For instance, the **np.arange** function generates a numpy array of incremented values. A numpy array (np.array) is a type of list that only contains numbers. 

Here are the np.arange function's arguments (which you can read about at the numpy documentation for [arange](https://numpy.org/doc/stable/reference/generated/numpy.arange.html)):

1. The beginning of the list (0)
2. The maximum value of the list (100; this is an exclusive maximum, so it will not be included in the list)
3. The incrementing value (1)

```python
# creating the t list
t = np.arange(0, 100, 1)
print("t list: \n", t)
```

As you can see above, a list of integers 0 - 99 were created. 

Now we want to multiply each value in the t list by the acceleration variable (9.8 m / s$^2$). Python's computations are very flexible, so all we need to do it multiply the list $t$ by the acceleration variable and add the initial velocity ($v_0$)! We are going to store it in the velocity variable $v$.

We'll assume that we're starting from rest, $v_0$=0 m/s. But feel free to try a different initial velocity and see how it changes.

$v(t) = a*t + v_0$

```python
# creating the new x-list (velocity)
a = 9.8 # acceleration due to gravity (m/s^2)
v_o = 0 # initial velocity (m/s)
xList = a * t +  v_o
print(xList)
```

Numpy lists (arrays) are different from typical lists. You can perform mathematical operations on a numpy array. 

```python
numpy_array = np.arange(5)
print(2*numpy_array)
regular_list = [0, 1, 2, 3, 4]
print(2*regular_list)
```

```python
numpy_array = np.arange(5)
print(numpy_array + 2)
regular_list = [0, 1, 2, 3, 4]
print(regular_list + 2)
```

Similarly you can create an array of values with a min and a max and a specific number of entries in the list with np.linspace.

The arguments are 
1. The beginning of the list (0)
2. The maximum value of the list (100)
3. The number of values to include in the list.

```python
# create a t list with linspace
t = np.linspace(0, 100, 100)
print(t)
```

Notice the differences between the t list made with np.arange and the one made by np.linspace. If you want to controll the spacing, use np.arange; if you want to control the number of elements use np.linspace!

```python
t = np.linspace(0, 100, 101)
print(t)
```

```python
# then create a new x-list (velocity) using that t-list
v = a*t
print(v)
```

## Make a Plot!

Now try plotting $v$ versus $t$ so that velocity is on the y-axis and time is on the x-axis. The arguments for plt.plot(x, y) are an array of x-values and an array of y-values. You can also use a list, but you have to make sure that all the values in the list are numbers (either integers or decimals.)

```python
# make a plot here
plt.plot(t,v)
```

## Labels
 
The graph above is nice, but it could be a lot more descriptive! We should add labels to this plot using the plt.xlabel and plt.ylabel methods! The arguments for each method is a string describing their coresponding axis. 

It is good practice to include an English-language word (i.e. not the name of variable) and units in parenthesis or brackets. This helps provide clarity in your work. As an example, the $t$ time variable should be labeled "Time (s)" rather than "t" or "s" or "sec".

```python
plt.plot(t, v)
plt.xlabel("Time (s)")
plt.ylabel("Velocity (m/s)")
```

## Plot Styles

Sometimes you want to make a plot that connects all the points in a line like we do above. Other times you want to make a plot with just the points. This is called a "scatter" plot. 

You can make a scatter plot either by changing the marker style or by using plt.scatter.

```python
# This sets the points to all "o"'s
plt.plot(t, v, 'o')
plt.xlabel("Time (s)")
plt.ylabel("Velocity (m/s)")

# So does this. But this time with a "square" marker
plt.plot(t, v, marker='s')
plt.xlabel("Time (s)")
plt.ylabel("Velocity (m/s)")
```

```python
# Here's how to use plt.scatter:
plt.scatter(t, v)
plt.xlabel("Time (s)")
plt.ylabel("Velocity (m/s)")
```

## Multiple Plots

What if we want to compare two different speeds on the same graph? Simply plot the two line graphs in the same cell! 

```python
# creation of v1 and v2 lists (two different velocities, )
t = np.arange(0, 100, 1)

# initial velocity
v10 = 0
# acceleration
acceleration1 = 10
v1 = acceleration1 * t + v10

# initial velocity
v20 = 100
# acceleration
acceleration2 = 20
v2 = acceleration2 * t + v20
```

Try playing with the initial velocities and accelerations above to see how it modifies the graph.

```python
# our new plot!
plt.plot(t, v1)
plt.plot(t, v2)

plt.xlabel("Time (s)")
plt.ylabel("Velocity (m/s)")
plt.title("Comparison of two velocities")
```

## Legends

Now that both line graphs are plotted, how do we distinguish one line graph from another? We can add a label each plot that describes what each line graph represents. 

To display a legend of these labels, include the plt.legend() function in the code cell.

```python
# plot the line graphs with labels
plt.plot(t, v1, label='velocity 1')
plt.plot(t, v2, label='velocity 2')

plt.xlabel("Time (s)")
plt.ylabel("Velocity (m/s)")
plt.title("Comparison of two velocities")

plt.legend()
```

## Data with Errorbars
Often our data has some uncertainty and we can use an errorbar plot to reflect that.

Let's start by making some data that oscillates sinusoidally, like the position, $y = A cos(\omega t)$, of a mass on a spring. We will use the [sin()](https://docs.scipy.org/doc/numpy/reference/generated/numpy.sin.html) function from numpy. 

```python
t = np.arange( 0, 100, 5) # seconds
amplitude = 1.0 # y-position of the mass, meters
omega = np.pi/30. # angular velocity omega, radians/second

y = amplitude * np.sin(omega * t)
```

Now let's assume that we can only measure the position of the mass on the spring to a certain precision, given by $\sigma$. So any value that we record will be uncertain by that number.

For instance, if I use a (not so great) ruler to measure the position of the mass on the spring that only has 0.2 m increments.

```python
sigma = 0.2 # precision of our ruler, meters
```

Now I can make a plot with error bars that include that uncertainty. Review the [matplotlib.errorbar](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.errorbar.html) documentation to learn more about the function we will use to make the plot.

```python
plt.errorbar(t, y, sigma)
```
