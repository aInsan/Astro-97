Make a long array of 10,000 Normally distributed numbers and name it data_arr

Then add 2 to every element in the array

```python
import matplotlib.pyplot as plt
import numpy as np

data_arr = np.random.normal(0,1,10000)

data_arr += 2
```

Let's take the square root of data_arr, but first let's make sure there are no negative numbers

First create a boolean array that is True wherever data_arr is less than zero

Then use that boolean array to change the value of each element in data_arr than is less than 0 to be 0.

(Hint: if you can assign a new value to an array element like this, arr[1] = 0 with a single index, can you do the same with an array of indexes or with a boolean array? )

Finally find the number of elements in data_arr that have a square root greater than 2

```python
bool_arr = data_arr < 0
data_arr[bool_arr] = 0

bool_sqrt_arr = data_arr**0.5 > 2
print(f"The number of elements with sqrt() greater than 2 is {np.sum(bool_sqrt_arr)}")
```

Create a structured array with a column for an angle that runs from 0 to 2 $\pi$

Then have the other columns be the sin, cos, and tan of the angle

```python
angle_dtype = np.dtype([('angle', float), ('sin(x)', float), ('cos(x)', float), ('tan(x)', float)])
num_elements = 480
angle_arr = np.zeros(num_elements, dtype=angle_dtype)
angle_arr['angle'] = np.linspace(0, 2*np.pi+0.00001, num_elements)
angle_arr['sin(x)'] = np.sin(angle_arr['angle'])
angle_arr['cos(x)'] = np.cos(angle_arr['angle'])
angle_arr['tan(x)'] = np.tan(angle_arr['angle'])
```

Print out each row in the structured array with 3 decimal points for each number. Start with a line that gives the name of each column

```python

print("|=============|=========|=========|=========|")
print("| angle (rad) |  sin(x) |  cos(x) |  tan(x) |")
print("|=============|=========|=========|=========|")

for i in range(num_elements):
    print(f"|{angle_arr['angle'][i]:^13.3f}|{angle_arr['sin(x)'][i]:^9.3f}\
|{angle_arr['cos(x)'][i]:^9.3f}|{angle_arr['tan(x)'][i]:^9.3f}|")
    
print("|=============|=========|=========|=========|")
```

```python
import matplotlib.pyplot as plt

plt.plot(angle_arr['angle'], angle_arr['sin(x)'], label='sin(x)')
plt.plot(angle_arr['angle'], angle_arr['cos(x)'], label='cos(x)', color='red')
plt.legend()
plt.show()
plt.plot(angle_arr['angle'], angle_arr['tan(x)'], label='tan(x)', color='green')
plt.ylim(-20,20)
plt.legend(loc='lower left')
plt.show()
```
