Luminosity measures the total energy per second generated by a star [Joules/sec]. Luminosity of a star is independent of the distance to that star. 

Since the ancient times, people have been using a system of ranking stars on the sky by their apparent brightness (apparent luminosity).  This system is called the Magnitude scale. The scale was devised so that the brightest star our eyes can see has 1 mag, while the faintest star our eyes can see is 6 mag. Notice that the larger magnitudes are dimmer! Magnitude scale is a logarithmic scale; a change of 5 in magnitude corresponds to a change of a factor of 100 in apparent brightness.


The observed (or apparent) magnitude, m, is the brightness of a star we observe from Earth, while the absolute magnitude, M, is the brightness observed from a distance of 10 pc from the object. (pc stands for parsec, a unit of distance where 1 pc = 3.26 light years).

The luminosity of the Sun is $L_{\odot}=3.83 \cdot 10^{26}$ Watts.

The apparent magnitude of the Sun is $m_{\odot}=-26.74$ mag, while the absolute magnitude is $M_{\odot}=4.83$
mag.

A general relationship between magnitudes and luminosities of two stars is:

$M_1-M_2 = - 2.5\log_{10} \dfrac{L_1}{L_2}$,

where M1 and M2 are the absolute magnitudes of objects, and L1 and L2 are the luminosities of the two respective objects.

First, lets solve $\boxed{100^{\frac{m - M}{5}} = \left(\frac{d}{10\,\mathrm{pc}}\right)^2}$ for m.

$ 100^{\frac{m - M}{5}} = \left(\frac{d}{10\,\mathrm{pc}}\right)^2 $

$\implies 10^{\frac{2}{5}(m - M)} = \frac{d^2}{100\,\mathrm{pc}} $

$\implies \log(10^{\frac{2}{5}(m - M)}) = \log(\frac{d^2}{100\,\mathrm{pc}}) = \log(d^2) - \log(100)$

$\implies \frac{2}{5}(m - M) = 2(\log(d) - 1) $

$\implies m - M = 5(\log(d) - 1) $

$\implies \boxed{M = m -  5(\log(d) - 1)} $

```python
import math #Importing the log10 function from the math module

#Using the derived formula for absolute magnitude define a function to calculate it from apparent magnitude and distance
def absolute_magnitude(mag,dist):
    '''
    Function to calculate the absolute magnitude of an object given its apparent magnitude and distance
    
    Inputs: mag - apparent magnitude of the object (unitless, float)
            dist - distance to the object in parsecs (parsec, float)
    '''

    return mag - 5*(math.log10(dist) - 1)
```

```python
#This List contains tuples for each objects with its name, apparent magnitude and distance
appMagTable = [("Full Moon",-12.5,1.246e-8),("New Moon",-2.5,1.246e-8),("Venus",-4.4,1.2315e-6),("Jupiter",-2.94,2.9338e-5),("Sirius",-1.5,2.64014),("Supernova 1987A",3.03,51509.0342)]


# Table Title
print("\n-- Table of Apparant Magnitude and Distance\n")

# Print table header
print(f"#{'='*17}#{'='*9}#{'='*15}#")
print(f"#{'Object [mag]':^17}#{'m [mag]':^9}#{'M [abs. mag]':^15}#")
print(f"#{'='*17}#{'='*9}#{'='*15}#")

# Create an absolute magnitude table (useful for later)
absMagTable = []

# Print data rows using a while loop
for body in appMagTable:
    absMagTable.append(absolute_magnitude(body[1],body[2]))
    print(f"# {(str(body[0])):<16}#{body[1]:^9}#{absolute_magnitude(body[1],body[2]):^15.6}#")
# close off table
print(f"#{'='*17}#{'='*9}#{'='*15}#")
```

The absolute magnitude of the sun is $M_{\odot}=4.83$
mag.

A general relationship between magnitudes and luminosities of two stars is:

$M_1-M_2 = - 2.5\log_{10} \dfrac{L_1}{L_2}$,


Since we want luminosity in terms of solar luminosity, we are solving for $\dfrac{L}{L_{\odot}}$

$M-M_{\odot} = - 2.5\log_{10} \dfrac{L}{L_{\odot}} $

$-0.4(M-M_{\odot}) = \log_{10} \dfrac{L}{L_{\odot}} $

$10^{-0.4(M-M_{\odot})} = \dfrac{L}{L_{\odot}} $

$\dfrac{L}{L_{\odot}} = 10^{-0.4(M-4.83)} $

```python
def luminosity_solar(mag):
    '''
    Function to calculate the luminosity (in terms of solar luminosity )of an object given its absolute magnitude
    
    Inputs: mag - absolute magnitude of the object (unitless, float)
    '''
    return 10**(-0.4*(mag - 4.83))

# Table Title
print("\n-- Table of Apparant Magnitude and Luminosity\n")

# Print table header
print(f"#{'='*17}#{'='*9}#{'='*15}#")
print(f"#{'Object [mag]':^17}#{'m [mag]':^9}#{'L [L_sol]':^15}#")
print(f"#{'='*17}#{'='*9}#{'='*15}#")



# Print data rows using a while loop
for body in appMagTable:
    print(f"# {(str(body[0])):<16}#{body[1]:^9}#{luminosity_solar(absolute_magnitude(body[1],body[2])):^15.6}#")
# close off table
print(f"#{'='*17}#{'='*9}#{'='*15}#")
```
