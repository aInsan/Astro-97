You have two different parameters (x,y) from a large dataset. There is inherent spread to each parameter, but in the dataset there are 2 populations. 

Create a plot that shows that there are two populations present in the data. 

Make sure your plot is:
- Easy to read
- Colorblind friendly 
- Properly labeled 

Use online resources and documentation to help you

Feel free to make plots for yourself to understand the data, just make it clear which is your final plot to present

```python
# Run this to generate your data
import numpy as np

x0 = 1.0
y0 = 2.0

x1 = -3.0
y1 = -2.0

# Number of data points from each population
N0 = int(1e6)
N1 = int(1e5)

x0_vals = np.random.normal(loc=x0, size=N0)
y0_vals = 10**np.random.normal(loc=y0, size=N0)

x1_vals = np.random.normal(loc=x1, size=N1)
y1_vals = 10**np.random.normal(loc=y1, size=N1)

# Your data to use
# Both populations together without labels
x_vals = np.append(x0_vals, x1_vals)
y_vals = np.append(y0_vals, y1_vals)
```

```python
import matplotlib.pyplot as plt

plt.scatter(x0_vals, y0_vals, marker='.', color='purple', label='Population 0', alpha=0.3)
plt.scatter(x1_vals, y1_vals, marker='^',color='green', label='Population 1', alpha=0.3)
plt.yscale('log')
plt.ylabel('y (log scale)')
plt.xlabel('x')
plt.legend()
plt.title('Scatter Plot of Two Populations')
plt.show()
```
