### Exercise 4.2 

Run the following cell to create a new list:

```python
a_list = [57.9, 108.2, 149.6, 227.9, 778.6, 1433.5, 2872.5, 4495.1]
```

Those are the semi-major axes of the planets in millions of km. Write a for loop that will create a new list of these distances in AU. For reference, 1 AU = 1.496e8 km. Print out this new list.

```python
au_list = []
AU = 1.496e8
for a in a_list:
    au_list.append(a*10**6/AU)

print(au_list)
```

Create a new list of the periods of these planets using a for loop. Print this new list as well. For reference, Kepler's law states that $P^2 = a^3$, where $P$ is the period in years, and $a$ is the semi-major axis in AU.

```python
periods = []
for dist in au_list:
    periods.append(np.sqrt(dist**3))

print(periods)
```

Venus is the 2nd planet from the Sun. Use indexing (e.g., name_of_list[index]) to calculate Venus' period in days. Now, write in a formatted print statement (lecture 02) the period of Venus in days. Give the value to two decimal places of precision. (Remember: one year is 365.25 days.)

```python
periodVenus = periods[1]
periodVenusDays = periodVenus * 365.25

print(f"The orbital period of Venus is {periodVenusDays:.2f} days.")
```

### Exercise 4.3

Back to plotting!

Let's plot the position of a mass on a spring as a function of time like we did in lecture 4, but this time lets plot it for 3 different frequencies all on the same plot. 

$y(t) = A*sin(\omega * t)$

You will use a for loop to iterate over the list of frequencies where inside the loop you will calculate the array of positions as a function of time, and plot it. 

Make sure to label each curve and include axis labels. 

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
from fractions import Fraction

Omegas = [np.pi/2, np.pi, 3*np.pi/2] # angular velocity omega, radians/second

t = np.linspace(0, 8, 100) # time, seconds

A = 5.0 # amplitude, meters

# for loop to iterate over Omegas
displacements = []
for omega in Omegas:
    displacements.append(A * np.sin(omega * t)) # displacement, meters

# make plot axis labels and legend
plots = []
plot_colors = ['r', 'g', 'b']
for i in range(len(displacements)):
    plt.plot(t, displacements[i], label=f"ω = {Fraction(Omegas[i]/np.pi):.2f}π", color = plot_colors[i])


# label the axes and title the plot
plt.xlabel("Time (s)")
plt.ylabel("Displacement (m)")
plt.title("Simple Harmonic Motion of 3 objects")

plt.xlim(0, 11) # scales the x axis so the legend doesn't overlap the plot
plt.ylim(-8.5, 8.5) # scales the y axis so the legend doesn't overlap the plot
plt.gca().set_aspect(0.1)
# show the legends
plt.legend()
```
